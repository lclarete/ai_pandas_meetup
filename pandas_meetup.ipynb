```python
# rename columns
def rename_cols(cols):
    # lower the cases using the string method lower()
    cols = cols.str.lower()
    # replace values to normalize the column's name
    cols = cols.str.replace(' ', '_').str.replace(')','')
    cols = cols.str.replace(':', '_').str.replace('(', '')
    cols = cols.str.replace('.', '_').str.replace('/', '')
    return(cols)
```

```python
# call the rename_cols() function with df.columns as an argument
df.columns = rename_cols(df.columns)
```

## Preprocessing

### One hot encoding

```python
pd.get_dummies(df.sex_rec).head()
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

df_cancer = pd.DataFrame(cancer['data'],
                         columns=cancer['feature_names'])
```

```python
from sklearn.preprocessing import StandardScaler

# importar biblioteca para normalizar os dados
scaler = StandardScaler()

# fitar o modelo de normalização de acordo com o padrão dos dados
scaler.fit(df_cancer)

# efetivamente normalizar os dados
scaled_data = scaler.transform(df_cancer)
```

```python
scaled_data
```

```python
# importar a função PCA da biblioteca sklearn
from sklearn.decomposition import PCA

# instanciar a biblioteca e definir o número de componentes
pca = PCA(n_components=2, random_state=42)

# treinar o modelo de acordo com os dados normalizados
pca.fit(scaled_data)

# transformar os dados usando a função transform
x_pca = pca.transform(scaled_data)

plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],
            # target variable
            c=cancer['target'],
            # colormap
            cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```
